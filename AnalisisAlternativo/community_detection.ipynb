```python
import networkx as nx
import matplotlib.pyplot as plt 
```

```python
graph = {
    1: [2, 3, 4, 5],
    2: [1, 3, 6, 7],
    3: [1, 2, 8, 9],
    4: [1, 10, 11, 12],
    5: [1, 13, 14, 15],
    6: [2, 16, 17, 18],
    7: [2, 19, 20, 21],
    8: [3, 22, 23, 24],
    9: [3, 25, 26, 27],
    10: [4, 28, 29, 30],
    11: [4],
    12: [4],
    13: [5],
    14: [5],
    15: [5],
    16: [6],
    17: [6],
    18: [6],
    19: [7],
    20: [7],
    21: [7],
    22: [8],
    23: [8],
    24: [8],
    25: [9],
    26: [9],
    27: [9],
    28: [10],
    29: [10],
    30: [10],
}

G = nx.Graph(graph)
```

```python
nx.draw_networkx(G)
```

```python
assignment = nx.community.louvain_communities(G, seed=123)
assignment
```

```python
ass = {i: j for j in range(len(assignment)) for i in assignment[j]}
color_map = [ass[node] for node in G.nodes()]

# Draw the graph with nodes colored by community
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, node_color=color_map, cmap=plt.cm.get_cmap('viridis'), with_labels=True, node_size=300)
plt.show()
```
